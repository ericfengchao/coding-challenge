```python
class Solution(object):
    def sliding_merge_sets(self, seqs, k):
        longest = 0
        merge = []
        
        k_remain = k
        i = 0
        while i < len(seqs):
            seq = seqs[i]
            i += 1
            
            if len(merge) == 0:
                merge.append(seq)
                longest = max(longest, seq[1] - seq[0] + k_remain)
                continue
                
            top = merge[-1]
            gap = seq[0] - top[1]
            
            if gap > k_remain:
                longest = max(longest, merge[-1][1] - merge[0][1]+k_remain)
                while len(merge) > 0:
                    dequeue = merge.pop(0)

                    if len(merge) > 0:
                        gap_returned = merge[0][0] - dequeue[1]
                        k_remain += gap_returned
                        if k_remain >= gap:
                            break
                    else:
                        gap = 0
                        k_remain = k
                        
            merge.append(seq)
            k_remain -= gap
            longest = max(longest, seq[1] - merge[0][0]+k_remain)
            
        return longest
            
    def characterReplacement(self, s, k):
        """
        :type s: str
        :type k: int
        :rtype: int
        """
        c_seqs = {}
        i = 0
        while i < len(s):
            c = s[i]
            if c not in c_seqs:
                c_seqs[c] = []
            j = i+1
            while j < len(s) and s[j] == c:
                j += 1
            c_seqs[c].append((i,j))
            i = j
        
        longest_seq = 0
        for c, seqs in c_seqs.items():
            longest_seq_per_c = self.sliding_merge_sets(seqs, k)
            print(c, seqs, longest_seq_per_c)
            longest_seq_per_c = min(len(s), longest_seq_per_c)
            longest_seq = max(longest_seq, longest_seq_per_c)
            
        return longest_seq
    
    def charReplace(self, s, k):
        count = {}
        res = 0

        l = 0
        maxf = 0
        for r in range(len(s)):
            count[s[r]] = 1 + count.get(s[r], 0)
            maxf = max(maxf, count[s[r]])
            print(count, maxf)

            if (r - l + 1) - maxf > k:
                count[s[l]] -= 1
                l += 1
                print('  ', count, r, l)

            res = max(res, r - l + 1)
        return res
```
